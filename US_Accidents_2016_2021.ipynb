```python
hour_of_accident=df.Start_Time.dt.hour
```

```python
sns.distplot(hour_of_accident,bins=24,kde=False,norm_hist=True)
```

```python
#Comparing the accident time distribution on Sunday and Monday
sunday_start_time=df.Start_Time[df.Start_Time.dt.dayofweek==6]
sns.distplot(sunday_start_time.dt.hour,bins=24,kde=False,norm_hist=True)
```

```python
monday_start_time=df.Start_Time[df.Start_Time.dt.dayofweek==0]
sns.distplot(monday_start_time.dt.hour,bins=24,kde=False,norm_hist=True)
```

-The hour of accident happening in sunday and mondays is very much different from each other ,on sundays the peek occurs between 10 am to 3pm ,unlike weekdays

```python
#Now checking the accident distribution in months 
sns.distplot(df.Start_Time.dt.month,bins=12,kde=False,norm_hist=True)
```

```python
 #accidents in each month of 2016
 df_2016=df.Start_Time[df.Start_Time.dt.year == 2016]
 sns.distplot(df_2016.dt.month,bins=12,kde=False,norm_hist=True)
```
